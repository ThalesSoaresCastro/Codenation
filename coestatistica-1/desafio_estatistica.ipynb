# Descrição do problema:

A resposta deve conter os valores da média, mediana, moda e desvio padrão da pontuação de crédito para cada estado do dataset.

**Separando os clientes por Estado**

```python
df['estado_residencia'].unique()
```

```python
data ={
        'SC':df[df['estado_residencia'] == 'SC'].pontuacao_credito,
        'RS':df[df['estado_residencia'] == 'RS'].pontuacao_credito,
        'PR':df[df['estado_residencia'] == 'PR'].pontuacao_credito,
}
```

```python
credito_estado = pd.DataFrame(data, columns=['SC', 'RS', 'PR'])
```

**Tratando dados faltantes**

* Verificando os dados faltantes de cada Coluna
* Substituir os dados faltantes de cada coluna pela média, mediana e moda.

```python
credito_estado.isnull().sum()
```

```python
credito_estado.head(20)
```

```python
def trata_dados(tipo):
    if tipo == 'media':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].mean()
        #Aplica a média da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].mean(), inplace=True)
    
    elif tipo == 'mediana':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].median()
        #Aplica a mediana da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].median(), inplace=True)
    
    elif tipo == 'moda':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].mode()[0]
        #Aplica a moda da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].mode()[0], inplace=True)
    
    #Substitui pelo quantil de 0.25.
    elif tipo == 'q25':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].quantile(q=0.25)
        #Aplica o quantil 0.25 da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].quantile(q=0.25), inplace=True)

    #Substitui pelo quantil de 0.50, que é o mesmo valor da mediana.
    elif tipo == 'q50':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].quantile(q=0.50)
        #Aplica o quantil 0.50 da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].quantile(q=0.50), inplace=True)
            
    #Substitui pelo quantil de 0.75.
    elif tipo == 'q75':
        #credito_estado[credito_estado['SC'].isnull()]=credito_estado['SC'].quantile(q=0.75)
        #Aplica o quantil 0.75 da coluna em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(credito_estado[e].quantile(q=0.75), inplace=True)

    elif tipo=='zero':
        #credito_estado[credito_estado['SC'].isnull()]=0
        #Aplica o valor 0(zero) em cada valor faltante da coluna.
        for e in credito_estado:
            credito_estado[credito_estado[e].isnull()]=credito_estado[e].fillna(0, inplace=True)

    else:
        print('Deve-se escolher uma das seguintes opções :\n *Média: media\n *Mediana: mediana\n *Moda: moda\n *Quantil 0.25: q25\n *Quantil 0.50: q50\n *Quantil 0.75: q75\n *Zero(0): zero\n')
    
```

```python
credito_estado = pd.DataFrame(data, columns=['SC', 'RS', 'PR'])
```

```python
#Função para calcular a moda, mediana, média e desvio padrão de cada estado.
def calc_estatistico():
    cal_est=[]
    for e in credito_estado:
        val = [credito_estado[e].mode()[0], credito_estado[e].median(), credito_estado[e].mean(), credito_estado[e].std()]
        print('Estado: ',e,'\nEstatisticas:','\n( moda,mediana,média e desvio padrão): ',val)
        cal_est.append(val)
    return cal_est
```

```python
est_data = calc_estatistico()
est_data
```

```python
#criando o objeto json e inserindo a estatistica de cada estado.
obj={
    'SC':{
        'moda':est_data[0][0],
        'mediana':est_data[0][1],
        'media':est_data[0][2],
        'desvio_padrao':est_data[0][3]
    },
    'RS':{
        'moda':est_data[1][0],
        'mediana':est_data[1][1],
        'media':est_data[1][2],
        'desvio_padrao':est_data[1][3]
    },
    'PR':{
        'moda':est_data[2][0],
        'mediana':est_data[2][1],
        'media':est_data[2][2],
        'desvio_padrao':est_data[2][3]
    }
}

obj
```

```python
#tranformando o objeto em json.
with open('submission.json', 'w') as json_file:
    json.dump(obj, json_file, indent=4)
```
